# Coursera Courses Score

A simple approach to try to give a score to the courses on Coursera based on their reviews

## Score

Calculates a score relating the average rating to the number of reviews as suggested in [this thread](https://math.stackexchange.com/questions/942738/algorithm-to-calculate-rating-based-on-multiple-reviews-using-both-review-score)

## Dataset

The dataset is available on [Kaggle](https://www.kaggle.com/septa97/100k-courseras-course-reviews-dataset/version/2)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('reviews_by_course.csv')
df = df.groupby('CourseId').agg(['count','mean'])
df.columns = ['Num', 'Avg']
```

```python
import math
def score(row):
    Q = 300 # num of reviews threshold
    p = row['Avg']
    q = row['Num']
    return (5*p/10) + 5*(1-math.exp((-1*q) / Q))
    
df['Score'] = df.apply(score, axis=1)
df.sort_values('Score', inplace=True, ascending=False)
df.head(50)
```
